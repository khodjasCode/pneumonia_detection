```python
path = Path('/kaggle/input/pnevmoniya')
path.ls()
```

```python
pnev_train = (path/'train'/'PNEUMONIA').ls().sorted()
pnev_train
```

```python
norm_train = (path/'train'/'NORMAL').ls().sorted()
norm_train
```

```python
example = norm_train[33]
example_o = Image.open(example)
example_o
```

```python
illness = DataBlock(
    blocks = (ImageBlock, CategoryBlock),
    get_items = get_image_files,
    splitter = RandomSplitter(valid_pct=0.1, seed=12),
    get_y = parent_label,
    item_tfms = Resize(224)
)
```

```python
dls = illness.dataloaders(path/'train')
```

```python
model = resnet34
```

```python
learn = vision_learner(dls, model, metrics=accuracy, pretrained=False)
learn.model.load_state_dict(torch.load('/kaggle/input/resnet34/resnet34-b627a593.pth', weights_only=True), strict=False)
```

```python
learn.fine_tune(2)
```

```python
learn.validate()[1]
```

```python
labels = []
for index, row in sample_sol.iterrows():
    image_path = Path(f"/kaggle/input/pnevmoniya/test/{row['id']}")
    img = PILImage.create(image_path)
    prediction = learn.predict(img)
    label = prediction[2][1].item()
    labels.append(label)
```

```python
labels = [1 if label > 0.5 else 0 for label in labels]
sample_sol['labels'] = labels
```
